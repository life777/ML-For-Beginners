```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model, model_selection

x, y = datasets.load_diabetes(return_X_y=True)

print(x.shape)
print(x[0])
```

```python
x2 = x[:, 2]
print(x2.shape)
print(x2[0])

x2 = x2.reshape((-1, 1))
print(x2.shape)
print(x2[0])
```

```python
# Split into training and testing sets
x_train, x_test, y_train, y_test = model_selection.train_test_split(x2, y, test_size=0.33)
print(x_train.shape)
print(x_test.shape)
```

```python
# create LinearRegression model and train it
model = linear_model.LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(y_pred.shape)
```

```python
plt.scatter(x_test, y_test, color="black")
plt.plot(x_test, y_pred, color="blue", linewidth=3)
plt.xlabel("BMI")
plt.ylabel("Diabetes")
plt.title("BMI to Diabetes")

plt.show()
```

```python
error = np.absolute(y_pred - y_test)
plt.scatter(x_test, error, color="red")
plt.show()
```
